```python
import pickle
with open('id2word.pkl', 'rb') as handle:
    id2word = pickle.load(handle)
```

```python
with open('corpus.pkl', 'rb') as handle:
    coupus = pickle.load(handle)
```

```python
from gensim.models import CoherenceModel
# Compute Coherence Score

with open('data_words_nostops.pkl', 'rb') as handle:
    data_words_nostops = pickle.load(handle)
```

```python
import gensim
lda_model6 = gensim.models.LdaMulticore(corpus=coupus,
                                       id2word=id2word,
                                       num_topics=6, 
                                       random_state=100,
                                       chunksize=100)
```

```python
with open('lda_model6.pkl', 'wb') as handle:
    pickle.dump(lda_model6, handle)
```

```python
coherence_model_lda = CoherenceModel(model=lda_model6, texts=data_words_nostops, dictionary=id2word, coherence='c_v')
coherence_lda6 = coherence_model_lda.get_coherence()
print('\nCoherence Score: ', coherence_lda6)
```

```python
perplexity6 = lda_model6.log_perplexity(coupus)
print(perplexity6)
```
